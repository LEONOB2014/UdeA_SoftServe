### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
# this file is tsv -> tab separated values
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep = '\t')
```

```python
chipo.head()
```

### Step 4. How many products cost more than $10.00?

```python
# Converting item_price to float
change_type = lambda x: float(x)
chipo.item_price = chipo.item_price.str.replace('$', '').apply(change_type)
chipo.head()
```

```python
mask = chipo.item_price > 10
chipo_over10 = chipo[mask]
chipo_over10
```

```python
chipo_over10.head()
```

```python
# There are some items that have different cost, given some ingredient choices, so the combination of item_name and choice description must be created
# to be able to filter unique items
#chipo['unique_item'] = chipo.apply(lambda row: row['item_name'] + row['choice_description'], axis=1)
chipo_over10.fillna('_', inplace=True)
chipo_over10['unique_item'] = chipo_over10['item_name'] + chipo_over10['choice_description']

chipo_over10.head()
```

```python
chipo_over10.unique_item.nunique() #-> There are 781 unique items with its ingredient choices that cost over $10
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
chipo_over10.loc[:,['unique_item', 'item_price']]
```

### Step 6. Sort by the name of the item

```python
chipo_over10.loc[:,['unique_item', 'item_price']].sort_values(['unique_item'])
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo_over10.loc[:,['unique_item', 'item_price']].sort_values(['item_price'], ascending=False).head(1)
```
